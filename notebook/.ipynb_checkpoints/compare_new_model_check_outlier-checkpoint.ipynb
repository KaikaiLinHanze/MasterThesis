```python
import tensorflow as tf
import sys, imageio, os
sys.path.append("/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project")
from models.model import *
from models.data_prepare import *
import keras
```

```python
def data(select_path):
    storage_path = "/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project/datasets"
    path = os.path.join(storage_path,select_path)
    im_path = os.path.join(path, "slices.png")
    label_path = os.path.join(path, "labels.npy")
    im = imageio.imread(im_path)
    with open(label_path, "rb") as f:
        label = np.load(f)
    label = np.expand_dims(label,1)
    return im, label
X, y = data("vary_width")
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X, y, test_size=0.3, random_state=42,shuffle=True)
```

```python
X_train,y_train = data("varying_lum_and_focus_train")
X_test,y_test = data("varying_lum_and_focus_test")
```

```python
X_train_combine = np.concatenate([X_train,X_train_new], axis=0)
y_train_combine = np.concatenate([y_train,y_train_new], axis=0)
X_test_combine  = np.concatenate([X_test,X_test_new], axis=0)
y_test_combine = np.concatenate([y_test,y_test_new], axis=0)
X_train_combine.shape,y_train_combine.shape,X_test_combine.shape,y_test_combine.shape
```

```python
X_train_combine = data_preparation(X_train_combine)
X_test_combine = data_preparation(X_test_combine)
```

```python
model_BO_combine = keras.models.load_model("/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project/models/CNN_combine_best_BO")
model_BO_84 = keras.models.load_model("/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project/models/BO_CNN_MAE_0.84_vary_lum_focus")
model_BO_115 = keras.models.load_model("/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project/models/BO_CNN_MAE_1.15_vary_lum_focus")
model_HB_combine_old = keras.models.load_model("/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project/models/CNN_combine_best_HB_1204")
model_HB_combine_new = keras.models.load_model("/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project/models/CNN_combine_best_HB_1205")
model_HB_87 = keras.models.load_model("/home/ipausers/lin/Desktop/Graduation/Kai/Graduation-Project/models/HB_CNN_MAE_0.87_vary_lum_focus")
model_felix = keras.models.load_model("/home/ipausers/lin/Desktop/Graduation/AMFtrack/amftrack/ml/models/default_model")
```

```python
fig = plt.figure(figsize=(10, 4))
e,f,dataset = 4,5,2
IMAGESIZE = 120
OUTPUTSIZE= 80
def center_crop(x, margin):
    return x[..., margin:-margin, :]

def center_crop_layer(IMAGESIZE,OUTPUTSIZE):
    return keras.layers.Lambda(lambda x: center_crop(x,(IMAGESIZE-OUTPUTSIZE)//2),name="center_crop")

def data_preparation(X_test):
    data_preprocessing = keras.Sequential([
        tf.keras.Input(shape=(IMAGESIZE, 1)),
        center_crop_layer(IMAGESIZE, OUTPUTSIZE)])
    return data_preprocessing(X_test)
sub5 = compare_different_model(model[e],  data_preparation(Xtest[dataset]), ytest[dataset], x_range=[0,14], set_title=model_title[e], total=2, plot_number=e-3)
sub6 = compare_different_model(model[f],  data_preparation(Xtest[dataset]), ytest[dataset], x_range=[0,14], set_title=model_title[f], total=2, plot_number=f-3)
plt.savefig("Compare model1.png")
```
